```python
import numpy as np
import pandas as pd
import pylab as pl  
import matplotlib
from scipy.optimize import curve_fit, minimize # bring methods of optimization and root finding
%pylab inline 
numpy.__version__
```

Chi-squared Distribution

```python
# Chi-squared Distribution
distributions = ['chisq']
mymean = 100

# chi sq distribution degree of freedom is its mean
df = mymean

# generate 99 numbers as size of 99 samples,and make them into integer
mysize = (2000 / (np.array(range(1, 100)))).astype(int)
# print (mysize, mysize.shape)

# md['chisq]'s value is used to be a list, but now we make it a dictionary
md['chisq'] = {}

for n in mysize:
    md['chisq'][n] = np.random.chisquare(df, size = n)
    
# make a another dict within the former one    
md['chisq']['means'] = {}

# make a figure, or more precisely, a frame
axchisq_mu_n = pl.figure(figsize=(10,6)).add_subplot(111)

for nn in md['chisq']:
    if not type(nn) == str: # type('mean') = str
        md['chisq']['means'][nn] = md['chisq'][nn].mean() # store means under mean dict
        #and plot it
        axchisq_mu_n.plot(nn, md['chisq']['means'][nn], 'o') #axes is 坐标系
        axchisq_mu_n.set_xlabel('sample size', fontsize=18)
        axchisq_mu_n.set_ylabel('sample mean', fontsize=18)
        axchisq_mu_n.set_title('Chi-squared Distribution', fontsize=18)
        axchisq_mu_n.plot([min(mysize), max(mysize)], [df, df], 'k')
        
# make a histogram
allmeans = list(md['chisq']['means'].values())

pl.figure(figsize=(10, 10))
pl.hist(allmeans,bins=30)
pl.xlabel('sample mean', fontsize = 18)
pl.ylabel('N', fontsize = 18)
```

Normal Distribution

```python
# Normal Distribution
mymean = 100
mysize = (2000 / (np.array(range(1,100)))).astype(int)
# print mysize

md['normal'] = {}
md['normal']['means'] ={}

for n in mysize:
    md['normal'][n] = np.random.normal(mymean,1,size = n)

# ready to plot
axnormal_mu_n = pl.figure(figsize=(10,6)).add_subplot(111)

for nn in md['normal']:
    if not type(nn) == str: # type('mean') = str
        md['normal']['means'][nn] = md['normal'][nn].mean() # store means under mean dict
        #and plot it
        axnormal_mu_n.plot(nn, md['normal']['means'][nn], 'o') #axes is 坐标系
        axnormal_mu_n.set_xlabel('sample size', fontsize=18)
        axnormal_mu_n.set_ylabel('sample mean', fontsize=18)
        axnormal_mu_n.set_title('Normal Distribution', fontsize=18)
        axnormal_mu_n.plot([min(mysize), max(mysize)], [df, df], 'k')
        
# make a histogram
allmeans = list(md['normal']['means'].values())

pl.figure(figsize=(10, 10))
pl.hist(allmeans,bins=30)
pl.xlabel('sample mean', fontsize = 18)
pl.ylabel('N', fontsize = 18)

```
