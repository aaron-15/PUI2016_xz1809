```python
# aquiring data by using urllib2
import pylab as pl
import json
import urllib2 as ulr
%pylab inline
```

```python
# make up to a complete url
url1 = "http://bustime.mta.info/api/siri/vehicle-monitoring.json?key="
mykey = "b10d0106-4b13-4d99-bb93-bea07080ac66"
url2 = "&VehicleMonitoringDetailLevel=calls&LineRef="
ref = "B52"
url = url1 + mykey + url2 + ref
```

```python
# acquiring data
response = urllib2.urlopen(url)
Data = response.read().decode('utf-8')
# use the json.loads method to obtain a dictionary representation of the responose string 
data_dict = json.loads(Data)
# print data_dict
```

```python
# index to check the number of bus on the road
buscheck = data_dict['Siri']['ServiceDelivery']['VehicleMonitoringDelivery'][0]['VehicleActivity']
# print buscheck
```
